```python
def describe_pets(animal_type,pet_name):
    """display some info about a pet"""
    print(f"\n i have a {animal_type.title()}")
    print(f" i have a {animal_type.title()}'s name as {pet_name.title()}")               #positional argument
```

```python
describe_pets('cat','snowbell')
```

```python
def describe_pets(animal_type,pet_type='bruno'):
    """display info about a pet"""
    print(f"\n i have a {animal_type.upper()}")
    print(f" my {animal_type.upper()}'s name is {pet_type.upper()}")          #default argument
```

```python
describe_pets('cat')
```

```python
describe_pets(pet_type='snowbell',animal_type='rabbit')   #keyword argument
```
